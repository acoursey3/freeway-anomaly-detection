```python
import pandas as pd
import ast
import warnings
warnings.simplefilter(action='ignore', category=pd.errors.SettingWithCopyWarning)
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.colors import LinearSegmentedColormap
import matplotlib.style as mplstyle
jet = plt.cm.jet
# Invert the colormap
colors = [jet(x) for x in np.linspace(1, 0.5, 256)]
green_to_red = LinearSegmentedColormap.from_list('GreenToRed', colors, N=256)
colors = [jet(x) for x in np.linspace(0.5, 1, 256)]
red_to_green = LinearSegmentedColormap.from_list('GreenToRed', colors, N=256)
data = pd.read_csv('today.csv')
data_lanes = data[data['lane_dict_text'].str.contains("Lane1") 
                  & data['lane_dict_text'].str.contains("Lane2") 
                  & data['lane_dict_text'].str.contains("Lane3")
                  & data['lane_dict_text'].str.contains("Lane4")]
import datetime
import pytz
def time_trans(dt_str):
# Define the date-time string
    # Convert the string to a datetime object in CDT
    dt = datetime.datetime.strptime(dt_str, '%Y-%m-%d %H:%M:%S')
#     cdt = pytz.timezone('America/Chicago')
#     dt = cdt.localize(dt)
    # Convert to UTC and get the Unix timestamp
    unix_timestamp = dt.astimezone(pytz.UTC).timestamp()

    return(unix_timestamp)
```

```python
%%time
# Function to extract speed for a specific lane from lane_dict
def extract_lane_speed(lane_dict_str, lane_num):
    # Convert the string representation of dictionary to actual dictionary
    lane_dict = ast.literal_eval(lane_dict_str)
    
    # Search for the lane and extract the speed
    for value in lane_dict.values():
        if f"Lane{lane_num}" in value[0] and len(value) > 1:
            return value[1]
    return None

def extract_lane_volume(lane_dict_str, lane_num):
    # Convert the string representation of dictionary to actual dictionary
    lane_dict = ast.literal_eval(lane_dict_str)
    
    # Search for the lane and extract the speed
    for value in lane_dict.values():
        if f"Lane{lane_num}" in value[0] and len(value) > 1:
            return value[2]
    return None

def extract_lane_occupancy(lane_dict_str, lane_num):
    # Convert the string representation of dictionary to actual dictionary
    lane_dict = ast.literal_eval(lane_dict_str)
    
    # Search for the lane and extract the speed
    for value in lane_dict.values():
        if f"Lane{lane_num}" in value[0] and len(value) > 1:
            return value[3]
    return None

# Use .loc to set the new columns
data_lanes.loc[:, 'lane1_speed'] = data_lanes['lane_dict_text'].apply(lambda x: extract_lane_speed(x, 1))
data_lanes.loc[:, 'lane2_speed'] = data_lanes['lane_dict_text'].apply(lambda x: extract_lane_speed(x, 2))
data_lanes.loc[:, 'lane3_speed'] = data_lanes['lane_dict_text'].apply(lambda x: extract_lane_speed(x, 3))
data_lanes.loc[:, 'lane4_speed'] = data_lanes['lane_dict_text'].apply(lambda x: extract_lane_speed(x, 4))

data_lanes.loc[:, 'lane1_volume'] = data_lanes['lane_dict_text'].apply(lambda x: extract_lane_volume(x, 1))
data_lanes.loc[:, 'lane2_volume'] = data_lanes['lane_dict_text'].apply(lambda x: extract_lane_volume(x, 2))
data_lanes.loc[:, 'lane3_volume'] = data_lanes['lane_dict_text'].apply(lambda x: extract_lane_volume(x, 3))
data_lanes.loc[:, 'lane4_volume'] = data_lanes['lane_dict_text'].apply(lambda x: extract_lane_volume(x, 4))

data_lanes.loc[:, 'lane1_occ'] = data_lanes['lane_dict_text'].apply(lambda x: extract_lane_occupancy(x, 1))
data_lanes.loc[:, 'lane2_occ'] = data_lanes['lane_dict_text'].apply(lambda x: extract_lane_occupancy(x, 2))
data_lanes.loc[:, 'lane3_occ'] = data_lanes['lane_dict_text'].apply(lambda x: extract_lane_occupancy(x, 3))
data_lanes.loc[:, 'lane4_occ'] = data_lanes['lane_dict_text'].apply(lambda x: extract_lane_occupancy(x, 4))
```
